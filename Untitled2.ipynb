```python
def preprocess_image(image):
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    cv2.imshow('Grayscale Image', gray_image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    
    blurred_image = cv2.GaussianBlur(gray_image, (5, 5), 0)
    cv2.imshow('Blurred Image', blurred_image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    
    threshold_image = cv2.adaptiveThreshold(blurred_image, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 2)
    cv2.imshow('Threshold Image', threshold_image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    
    return gray_image
```

```python
import language_tool_python
import requests

my_tool = language_tool_python.LanguageTool('en-US')
def levenshtein(s1, s2):
    if len(s1) < len(s2):
        return levenshtein(s2, s1)

    if len(s2) == 0:
        return len(s1)

    previous_row = range(len(s2) + 1)
    for i, c1 in enumerate(s1):
        current_row = [i + 1]
        for j, c2 in enumerate(s2):
            insertions = previous_row[j + 1] + 1
            deletions = current_row[j] + 1       # than s2
            substitutions = previous_row[j] + (c1 != c2)
            current_row.append(min(insertions, deletions, substitutions))
        previous_row = current_row

    return previous_row[-1]

def spelling_accuracy(extracted_text):
    spell_corrected = TextBlob(extracted_text).correct()
    return ((len(extracted_text) - (levenshtein(extracted_text, spell_corrected)))/(len(extracted_text)+1))*100
    
def grammatical_accuracy(extracted_text):
    try:
        spell_corrected = TextBlob(extracted_text).correct()
        correct_text = my_tool.correct(spell_corrected)
        extracted_text_set = set(spell_corrected.split(" "))
        correct_text_set = set(correct_text.split(" "))
        n = max(len(extracted_text_set - correct_text_set),
                len(correct_text_set - extracted_text_set))
        return ((len(spell_corrected) - n) / (len(spell_corrected) + 1)) * 100
    except Exception as e:
        print("Error while calculating grammatical accuracy:", e)
        return 0  # Return 0 accuracy in case of error

spelling_accurac = spelling_accuracy(extracted_text)
grammatical_accurac = gramatical_accuracy(extracted_text)
print("Spelling Accuracy: ", spelling_accurac)
print("Grammatical Accuracy: ", grammatical_accurac)
percentage_correction = (100 - spelling_accurac + 100 - grammatical_accurac) 
print("Percentage of corrections:", percentage_correction)
```

```python
extracted_text="Unless we can get e definite tne on him its hopelss....SSRERMB wit turn up one day, yall se} Reddy said contdently. IP we plug away at He leads we have...""IP only the police would do something .. Hugh Cried out in exasperation. Tie put Crerything E kro befare them..."
import language_tool_python
from textblob import TextBlob
import requests

my_tool = language_tool_python.LanguageTool('en-US')

def needleman_wunsch(s1, s2, gap_penalty=-1, mismatch_penalty=-1, match_reward=1):
    n = len(s1)
    m = len(s2)

    matrix = [[0] * (m + 1) for _ in range(n + 1)]

    for i in range(1, n + 1):
        matrix[i][0] = matrix[i - 1][0] + gap_penalty
    for j in range(1, m + 1):
        matrix[0][j] = matrix[0][j - 1] + gap_penalty

    for i in range(1, n + 1):
        for j in range(1, m + 1):
            if s1[i - 1] == s2[j - 1]:
                diagonal_score = matrix[i - 1][j - 1] + match_reward
            else:
                diagonal_score = matrix[i - 1][j - 1] + mismatch_penalty
            horizontal_score = matrix[i][j - 1] + gap_penalty
            vertical_score = matrix[i - 1][j] + gap_penalty
            matrix[i][j] = max(diagonal_score, horizontal_score, vertical_score)

    return matrix[n][m]

def spelling_accuracy(extracted_text):
    spell_corrected = TextBlob(extracted_text).correct()
    return ((len(extracted_text) - needleman_wunsch(extracted_text, spell_corrected))/(len(extracted_text)+1))*100
    
def grammatical_accuracy(extracted_text):
    try:
        spell_corrected = TextBlob(extracted_text).correct()
        correct_text = my_tool.correct(spell_corrected)
        extracted_text_set = set(spell_corrected.split(" "))
        correct_text_set = set(correct_text.split(" "))
        n = max(len(extracted_text_set - correct_text_set),
                len(correct_text_set - extracted_text_set))
        return ((len(spell_corrected) - n) / (len(spell_corrected) + 1)) * 100
    except Exception as e:
        print("Error while calculating grammatical accuracy:", e)
        return 0

spelling_accurac = spelling_accuracy(extracted_text)
print("Spelling Accuracy: ", spelling_accurac)
# grammatical_accurac = gramatical_accuracy(extracted_text)
# print("Grammatical Accuracy: ", grammatical_accurac)
percentage_correction = (100 - spelling_accurac) 
print("Percentage of corrections:", percentage_correction)
```

```python
def process_images(directory, label):
    data = []
    for filename in os.listdir(directory):
        if filename.endswith(".jpg") or filename.endswith(".png"):  # Assuming images are jpg or png format
            image_path = os.path.join(directory, filename)
            image = cv2.imread(image_path)
            preprocessed_image = preprocess_image(image)
            extracted_text = pytesseract.image_to_string(preprocessed_image)
            spelling_acc = spelling_accuracy(extracted_text)
            grammatical_acc = grammatical_accuracy(extracted_text)
            percentage_correction = (100 - spelling_acc + 100 - grammatical_acc)
            data.append([spelling_acc, grammatical_acc, percentage_correction, label])
    return data

dyslexic_data = process_images(image_directory_dyslexic, 1)

non_dyslexic_data = process_images(image_directory_non_dyslexic, 0)

with open('data.csv', mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['Spelling Accuracy', 'Grammatical Accuracy', 'Percentage of Correction', 'Label'])
    writer.writerows(dyslexic_data)
    writer.writerows(non_dyslexic_data)

print("Data has been written to data.csv file.")
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('filtered_data.csv')
```

```python
print("Data")
data.head()
```

```python
X = data.drop(columns=['Label'])
y = data['Label']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.27, random_state=42)
```

```python
rf_classifier = RandomForestClassifier(n_estimators=1000, random_state=42, min_samples_split=3)
rf_classifier.fit(X_train, y_train)
```

```python
y_pred = rf_classifier.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy of Random Forest classifier:", accuracy)
```

```python
conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(conf_matrix)
```
